```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page containing the list of movies
url = "https://en.wikipedia.org/wiki/List_of_films_set_in_the_future"

# Fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Example to find tables in the Wikipedia page
tables = soup.find_all("table", {"class": "wikitable"})

# Load the first table as a pandas DataFrame
movies_df = pd.read_html(str(tables[0]))[0]

# Preview the DataFrame
movies_df
```

```python
import requests

# Your OMDb API key
api_key = 'a2c946bf'

def get_movie_details(title):
    url = f"http://www.omdbapi.com/?t={title}&apikey={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if data['Response'] == 'True':
            imdb_score = data.get('imdbRating', 'N/A')
            runtime = data.get('Runtime', 'N/A')
            return imdb_score, runtime
    return 'N/A', 'N/A'

# Example usage
title = "Inception"
imdb_score, runtime = get_movie_details(title)
print(f"Title: {title}, IMDb Score: {imdb_score}, Runtime: {runtime}")
```

```python
# Initialize new columns
movies_df['IMDb Score'] = ''
movies_df['Runtime'] = ''

for index, row in movies_df.iterrows():
    title = row['Title']  # Adjust based on your DataFrame column name
    imdb_score, runtime = get_movie_details(title)
    movies_df.at[index, 'IMDb Score'] = imdb_score
    movies_df.at[index, 'Runtime'] = runtime

# Preview the updated DataFrame
movies_df
```
